```python
#imports 
import pandas as pd

```

```python
# User data
user_names = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table('Data/users.dat', sep="::", header=None, engine='python', encoding='ISO-8859-1', names=user_names)


#Rating data
rating_names = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('Data/ratings.dat', sep='::', header=None, engine='python', encoding='ISO-8859-1', names=rating_names)

#movie data
movie_names = ['movie_id', 'title', 'genres']
movies = pd.read_table('Data/movies.dat', sep="::", header=None, engine='python', encoding='ISO-8859-1', names=movie_names)
```

Using pandas's merge function to merge ratings with users and finally merging the results with movies data.

```python
merged_data = pd.merge(pd.merge(ratings, users), movies)
merged_data
```

The use of .loc function

```python
merged_data.loc[(merged_data.gender == 'M') & (merged_data.rating > 4)]
```

Using pivot_table to get the mean rating.
index is used in place of rows because rows became deprecated

```python
mean_ratings = merged_data.pivot_table('rating', index='title', columns='gender', aggfunc='mean')
mean_ratings.head()
```

Grouping data using size()

```python
ratings_by_title = merged_data.groupby('title').size()
ratings_by_title.head()

```

Filtering

```python
active_titles = ratings_by_title.index[ratings_by_title >= 250]
active_titles
```

```python
active_titles = ratings_by_title[ratings_by_title >= 250]
active_titles.head()
```

```python
mean_ratings = mean_ratings.iloc[active_titles]
mean_ratings
```
